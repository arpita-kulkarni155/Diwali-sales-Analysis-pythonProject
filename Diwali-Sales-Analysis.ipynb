## Exploratory Data Analysis

## Gender

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df,palette='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gender
```

```python
bx=sns.barplot(data=sales_gender,x='Gender',y='Amount',palette='Set1')
```

#### From above we can see that most of the buyers are females and even the purchasing power of women are greater than men

## Age

```python
df.columns
```

```python
bx=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in bx.containers:
    bx.bar_label(bars)
```

```python
sales_AgeGroup=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_AgeGroup
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_AgeGroup,palette='deep')
```

#### From above graph we can see that the most of the buyers are of age group between 26-35yrs Female

## State 

```python
df.columns
```

```python
# statewise total no of orders (top 10 states)
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sales_state
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=sales_state,x='State',y='Orders',palette='colorblind')
```

```python
# ststewise total amount (top 10 state)
state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
state_amount
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=state_amount,x='State',y='Amount',palette='pastel')
```

#### From above Graph we can see that most of the orders and amount spent/total sales are Uttar Pradesh,Maharashtra,Karnataka respectively.

### Marital_Status

### Occupation

```python
plt.figure(figsize=(15,5))
ax=sns.countplot(x='Occupation',data=df,palette='muted')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_Occupation
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=sales_Occupation,x='Occupation',y='Amount',palette='Accent')
```

#### From this graph we can see that most of the buyers are from IT sector, Heathcare, Aviation

### Product Category

```python
plt.figure(figsize=(25,5))
ax=sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ProductCateg=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
```

```python
sales_ProductCateg
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=sales_ProductCateg,x='Product_Category',y='Amount')
```

#### From above graph we can see that most of the sold products are from Food,Clothing & Apparel,Electronics & Gadget category
